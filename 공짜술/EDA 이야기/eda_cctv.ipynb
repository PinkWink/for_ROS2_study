```python
def drawGraph():
    plt.figure(figsize=(10,6))
    plt.plot(t, np.sin(t), label='싸인')
    plt.plot(t, np.cos(t), label='코싸인')
    plt.grid()
    plt.legend()
    plt.xlabel('시간')
    plt.ylabel('크기')
    plt.title('싸인코싸인 곡선')
    plt.show()
```

```python
import numpy as np

fp1 = np.polyfit(data_result['인구수'], data_result['총계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
```

```python
f1(400000)
```

```python
fx = np.linspace(100000, 700000, 100)
```

```python
fx
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = data_result['총계'] - f1(data_result['인구수'])

```
